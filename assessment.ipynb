**Data Summary**

Brazilian E-Commerce Public Dataset by Olist

 This is a Brazilian ecommerce public dataset of orders made at Olist Store. The dataset has information of 100k orders from 2016 to 2018 made at multiple marketplaces in Brazil. Its features allows viewing an order from multiple dimensions: from order status, price, payment and freight performance to customer location, product attributes and finally reviews written by customers. They also released a geolocation dataset that relates Brazilian zip codes to lat/lng coordinates.


An order might have multiple items.
Each item might be fulfilled by a distinct seller.
All text identifying stores and partners where replaced by the names of Game of Thrones great houses.**

1. Customers Dataset
- This dataset has information about the customer and its location. Use it to identify unique customers in the orders dataset and to find the orders delivery location.

- At this system each order is assigned to a unique customer_id. This means that the same customer will get different ids for different orders. The purpose of having a customer_unique_id on the dataset is to allow you to identify customers that made repurchases at the store. Otherwise you would find that each order had a different customer associated with.

2. Geolocation Dataset
- This dataset has information Brazilian zip codes and its lat/lng coordinates. Use it to plot maps and find distances between sellers and customers.

3. Order Items Dataset
- This dataset includes data about the items purchased within each order

4. Payments Dataset
- This dataset includes data about the orders payment options.

5. Order Reviews Dataset
- This dataset includes data about the reviews made by the customers.

6. Order Dataset
- This is the core dataset. From each order you might find all other information.

7. Products Dataset
- This dataset includes data about the products sold by Olist.

8. Sellers Dataset
- This dataset includes data about the sellers that fulfilled orders made at Olist. Use it to find the seller location and to identify which seller fulfilled each product.

9. Category Name Translation
- Translates the product_category_name to english.



**Description of each columns of the dataset**

Brazilian E-Commerce Public Dataset
1. customers
- customer_id – Unique ID for a customer in a particular order. Each order gets its own customer_id, so the same   person may have multiple customer_ids. Used to connect orders to customers.
- customer_unique_id – Real customer ID that stays the same across all orders made by that person. Crucial for tracking repeat purchases, retention, and loyalty.
- customer_zip_code_prefix – First 5 digits of the customer’s postal code. Helps in mapping customers geographically, calculating distance to sellers, and regional analysis.
- customer_city – The city where the customer lives. Useful for city-level sales, clustering customers, and mapping performance.
- customer_state – Customer’s state abbreviation. Useful for state-level analysis, regional marketing campaigns, and logistics planning.

2. geolocations
- geolocation_zip_code_prefix – Zip code prefix. Links to Customers and Sellers.
- geolocation_lat – Latitude coordinate. Useful for mapping and distance calculation.
- geolocation_lng – Longitude coordinate. Useful for mapping and distance calculation.
- geolocation_city – City name. Useful for regional analysis.
- geolocation_state – State abbreviation. Useful for regional analysis.

3. order items
- order_id – Links the item to a specific order. Unique ID of the order, Think  of it as ONE SHOPPING BILL NUMBER.
- order_item_id – Sequence number of the item within the order. Helps count the number of items per order.
- product_id – ID of the product sold in this item. Links to the Products dataset.
- seller_id – ID of the seller fulfilling this product. Links to the Sellers dataset.
- shipping_limit_date – The latest date the seller can ship the product. Helps track SLA compliance.
- price (R$) – Price of the product. Useful for revenue, profitability, and pricing analysis. Does not include freight.
- freight_value (R$) – Shipping cost for this item. Useful for logistics and cost analysis.

4. order payments
- order_id – Links payment to a specific order.
- payment_sequential – Sequence number if multiple payments were made for the same order.
- payment_type – Method of payment (e.g., credit_card, boleto, voucher, debit_card). Useful for understanding - customer payment preferences.
- payment_installments – Number of installments paid. Shows whether customers prefer installments or full payment.
- payment_value – Amount paid in this payment transaction. Used to calculate total revenue per order.

5. order reviews
- review_id – Unique ID for the review. Primary key for this table
- order_id – The order that this review is about.
- review_score – Rating given by the customer (1 to 5). Key indicator of customer satisfaction.
- review_comment_title – Optional title of the review. Provides short feedback context.
- review_comment_message – Full text of the review. Useful for sentiment analysis.
- review_creation_date – Date when the review was submitted. Helps analyze feedback over time.
- review_answer_timestamp – Date when the seller or platform responded. Helps track response speed.

6. orders
- order_id – Unique identifier for each order. The main key to link all order-related tables.
- customer_id – ID of the customer who placed the order. Links to the Customers dataset.
- order_status – Current status of the order (e.g., delivered, shipped, canceled). Shows fulfillment, delays, and cancellations.
- order_purchase_timestamp – Date and time when the order was placed. Useful for sales trends, time-series analysis, and seasonality.
- order_approved_at – Date and time when payment was approved. Helps track payment processing speed and detect delays.
- order_delivered_carrier_date – Date when the carrier picked up the order. Marks the start of shipping.
- order_delivered_customer_date – Date when the customer actually received the order. Helps calculate delivery times and delays.
- order_estimated_delivery_date – Promised delivery date. Can be used to calculate whether the order was delivered on time or late.

7. products
- product_id – Unique ID of the product. Links to Order Items. Can help to see which product was sold in which order.
- product_category_name – Product category in Portuguese. Can be translated for reporting.
- product_name_lenght – Number of characters in the product name. Helps check completeness of product info.
- product_description_lenght – Number of characters in the product description. Helps check information quality.Can indicate if longer description helps in the growth of e-commerce.
- product_photos_qty – Number of photos uploaded for the product. Indicates quality of product listing.Can indicate if no. of photos helps in more sales
- product_weight_g – Weight of product in grams. Important for shipping cost calculation.
- product_length_cm – Length in cm. Important for shipping dimensions.
- product_height_cm – Height in cm. Important for shipping dimensions.
- product_width_cm – Width in cm. Important for shipping dimensions.

8. sellers
- seller_id – Unique ID of the seller. Links to Order Items.
- seller_zip_code_prefix – Postal code prefix of the seller. Used to calculate delivery distance.
- seller_city – City where the seller is located. Useful for regional performance and logistics analysis.
- seller_state – State where the seller is located. Useful for regional segmentation.

9. category translations
- product_category_name – Original category name in Portuguese.
- product_category_name_english – Translated category name in English. Used for reports and dashboards.
